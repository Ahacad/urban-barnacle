### 通用函数

```python
# treemap 
def draw_treemap(df1, colname="行业2", fontsize=10):
    mpl.rcParams.update({"font.size": fontsize})
    
    sizes = df1.groupby(colname).size()
    norm = matplotlib.colors.Normalize(vmin=min(sizes), vmax=max(sizes))
    colors = [mpl.cm.Blues(norm(value)) for value in sizes]

    squarify.plot(sizes=df1.groupby(colname).size(), label=df1.groupby(colname).groups.keys(), alpha=.8, color=colors)
```

```python
# pie chart
def draw_piechart(df1, colname="行业2", fontsize=10):
    plt.rcParams['font.size'] = fontsize
    
    names = df1.groupby(colname).size().keys()
    plt.pie(df1.groupby(colname).size(), labels=names, labeldistance=1.4)
```

## 创业板
